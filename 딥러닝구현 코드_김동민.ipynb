## 딥러닝 구현 기초
- 데이터 샘플이나 layer 자유롭게 활용
- layer 추가 및, 학습률과 반복 횟수 조정

```python
import math
import random

# 딥러닝 학습 시 사용할 변수 설정
a = random.uniform(-1,1)
b = random.uniform(-1,1)
c = random.uniform(-1,1)
d = random.uniform(-1,1)
e = random.uniform(-1,1)
f = random.uniform(-1,1)
g = random.uniform(-1,1)
h = random.uniform(-1,1)
i = random.uniform(-1,1)
j = random.uniform(-1,1)
k = random.uniform(-1,1)
l = random.uniform(-1,1)
m = random.uniform(-1,1)
n = random.uniform(-1,1)
o = random.uniform(-1,1)

# layer 추가로 인한 변수 설정 
p = random.uniform(-1, 1)
q = random.uniform(-1, 1)
r = random.uniform(-1, 1)
s = random.uniform(-1, 1)
t = random.uniform(-1, 1)
u = random.uniform(-1, 1)

# 학습률 설정 (0.00001에서 0.000001로 변경)
eta = 0.000001

# 데이터 및 label 설정
data = [(5, 10), (7, 18), (10, 8), (15, 15), (20, 3), (23, 12)]
labels = [-1, -1, +1, +1, -1, -1]

# activation function : sigmoid 설정
def sigmoid(x):
    return 1.0/(1.0+math.exp(-x))

# 각 layer에서 출력될 값의 함수 설정
def node_output(x, y, a, b, c):
    return a*x+b*y+c

# 역전파 과정에서 활용되는 sigmoid 함수 설정
def back_sigmoid(fore_gredient, para, this_value):
    return fore_gredient*para*sigmoid(this_value)*(1.0 - sigmoid(this_value))
```

```python
# 학습을 위한 반복문 설정 : 학습 횟수 150만번으로 설정
for counter in range(1500000):
    da=db=dc=dd=de=df=dg=dh=di=dj=dk=dl=dm=dn=do=dp=dq=dr=ds=dt=du=0.0
    
    for idx in range(len(data)):
        x=float(data[idx][0])
        y=float(data[idx][1])
        
        # 1 layer에 대한 계산
        z1 = node_output(x, y, a, b, c)
        z2 = node_output(x, y, d, e, f)
        X1 = sigmoid(z1)
        Y1 = sigmoid(z2)
        
        # 2 layer에 대한 계산
        z3 = node_output(X1, Y1, g, h, i)
        z4 = node_output(X1, Y1, j, k, l)
        X2 = sigmoid(z3)
        Y2 = sigmoid(z4)
        
        # 추가된 layer에 대한 계산
        z5 = node_output(X2, Y2, m, n, o)
        z6 = node_output(X2, Y2, p, q, r)
        X3 = sigmoid(z5)
        Y3 = sigmoid(z6)
        
        # 최종 layer에 대한 계산
        z = node_output(X3, Y3, s, t, u)
        
        Z = labels[idx]
        dz = 2*(z-Z)
        
        # 최종 layer에 대한 역전파 계산
        dz5 = back_sigmoid(dz, s, z5)
        dz6 = back_sigmoid(dz, t, z6)
        ds += dz*X3
        dt += dz*Y3
        du += dz
        
        # 추가된 layer에 대한 역전파 계산
        dz3 = back_sigmoid(dz5, m, z3) + back_sigmoid(dz6, p, z3)
        dz4 = back_sigmoid(dz5, n, z4) + back_sigmoid(dz6, q, z4)
        dm += dz5 * X2
        dn += dz5 * Y2
        do += dz5
        dp += dz6 * X2
        dq += dz6 * Y2
        dr += dz6
        
        # 2 layer에 대한 역전파 계산
        dz1 = back_sigmoid(dz3, g, z1) + back_sigmoid(dz4, j, z1)
        dz2 = back_sigmoid(dz3, h, z2) + back_sigmoid(dz4, k, z2)
        dg += dz3 * X1
        dh += dz3 * Y1
        di += dz3
        dj += dz4 * X1
        dk += dz4 * Y1
        dl += dz4
        
        # 1 layer에 대한 역전파 계산
        da += dz1 * x
        db += dz1 * y
        dc += dz1
        dd += dz2 * x
        de += dz2 * y
        df += dz2
    
    
    a -= eta*da
    b -= eta*db
    c -= eta*dc
    d -= eta*dd
    e -= eta*de
    f -= eta*df
    g -= eta*dg
    h -= eta*dh
    i -= eta*di
    j -= eta*dj
    k -= eta*dk
    l -= eta*dl
    m -= eta*dm
    n -= eta*dn
    o -= eta*do
    p -= eta*dp
    q -= eta*dq
    r -= eta*dr
    s -= eta*ds
    t -= eta*dt
    u -= eta*du
        
print(a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u) 
```

```python
def print_z(x, y):
    # 1 layer에 대한 계산
    A = sigmoid(node_output(x, y, a, b, c))
    B = sigmoid(node_output(x, y, d, e, f))
    
    # 2 layer에 대한 계산
    C = sigmoid(node_output(A, B, g, h, i))
    D = sigmoid(node_output(A, B, j, k, l))
    
    # 추가된 layer에 대한 계산
    E = sigmoid(node_output(C, D, m, n, o))
    F = sigmoid(node_output(C, D, p, q, r))
    
    # 최종 결과 계산
    z = node_output(E, F, s, t, u)
    return print(z)

print_z(5,10)
print_z(7,18)
print_z(10,8)
print_z(15,15)
print_z(20,3)
print_z(23,12)

# 모두 음성으로 표현, layer수를 늘려서 overfitting이 발생한 것으로 추정
```

## 응용문제 : 자유 데이터 활용 (Iris)
- 앞서 활용한 방식을 Iris 데이터에도 적용해, 동일한 방식으로 overfitting이 발생하는지 확인

```python
import math
import random

# 1 layer에 대한 변수 설정
a_ = random.uniform(-1,1)
b_ = random.uniform(-1,1)
c_ = random.uniform(-1,1)
d_ = random.uniform(-1,1)
e_ = random.uniform(-1,1)
f_ = random.uniform(-1,1)
g_ = random.uniform(-1,1)
h_ = random.uniform(-1,1)
i_ = random.uniform(-1,1)
j_ = random.uniform(-1,1)

# 2 layer에 대한 변수 설정
g = random.uniform(-1,1)
h = random.uniform(-1,1)
i = random.uniform(-1,1)
j = random.uniform(-1,1)
k = random.uniform(-1,1)
l = random.uniform(-1,1)

# 3 layer에 대한 변수 설정
m = random.uniform(-1,1)
n = random.uniform(-1,1)
o = random.uniform(-1,1)
p = random.uniform(-1,1)
q = random.uniform(-1,1)
r = random.uniform(-1,1)
s = random.uniform(-1,1)
t = random.uniform(-1,1)
u = random.uniform(-1,1)

# 앞선 학습과 동일한 환경 설정
eta = 0.000001

def sigmoid(x):
    return 1.0/(1.0+math.exp(-x))

def node_output(x, y, a, b, c):
    return a*x+b*y+c

def node_output2(x1, x2, x3, x4, a, b, c, d, e):
    return a*x1+b*x2+c*x3+d*x4+e

def back_sigmoid(fore_gredient, para, this_value):
    return fore_gredient*para*sigmoid(this_value)*(1.0 - sigmoid(this_value))
```

```python
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris()
data = iris.data
labels = iris.target

labels = np.expand_dims(labels, axis=1)
iris = np.concatenate((data, labels), axis=1)

iris = iris[(iris[:,4]==0.0) | (iris[:,4]==1.0)] # 2진 분류를 위해 0, 1 데이터만  사용

np.random.shuffle(iris)
data = iris[:,:4]
labels = iris[:,4:].reshape((-1,))

test_data = data[-20:]
test_labels = labels[-20:]
data = data[:-20]
labels = labels[:-20]
print('total data len:',len(iris.data), ', train len:', len(data), ', test len:',len(test_data))
```

```python
# 반복 수 변경
for counter in range(1500000):
    da_=db_=dc_=dd_=de_=df_=dg_=dh_=di_=dj_=0.0
    dg=dh=di=dj=dk=dl=dm=dn=do=dp=dq=dr=ds=dt=du=0.0
    if(counter % 100000 == 0):print(counter, end=' ', flush=True)
        
    for idx in range(len(data)):
        x1=float(data[idx][0])
        x2=float(data[idx][1])
        x3=float(data[idx][2])
        x4=float(data[idx][3])
        
        # 1 layer에 대한 계산
        z1 = node_output2(x1, x2, x3, x4, a_, b_, c_, d_, e_)
        z2 = node_output2(x1, x2, x3, x4, f_, g_, h_, i_, j_)
        X1 = sigmoid(z1)
        Y1 = sigmoid(z2)
        
        # 2 layer에 대한 계산
        z3 = node_output(X1, Y1, g, h, i)
        z4 = node_output(X1, Y1, j, k, l)
        X2 = sigmoid(z3)
        Y2 = sigmoid(z4)
                
        # 추가된 layer에 대한 계산
        z5 = node_output(X2, Y2, m, n, o)
        z6 = node_output(X2, Y2, p, q, r)
        X3 = sigmoid(z5)
        Y3 = sigmoid(z6)
        z = node_output(X3, Y3, s, t, u)
        
        Z = labels[idx]
        dz = 2*(z-Z)
        ds += dz * X3
        dt += dz * Y3
        du += dz

        # 추가된 layer에 대한 역전파 계산
        dz5 = back_sigmoid(dz, s, z5)
        dz6 = back_sigmoid(dz, t, z6)
        dm += dz5 * X2
        dn += dz5 * Y2
        do += dz5
        dp += dz6 * X2
        dq += dz6 * Y2
        dr += dz6
        
        # 2 layer에 대한 역전파 계산
        dz3 = back_sigmoid(dz, m, z3) + back_sigmoid(dz6, p, z3)
        dz4 = back_sigmoid(dz, n, z4) + back_sigmoid(dz6, q, z4)
        dg += dz3 * X1
        dh += dz3 * Y1
        di += dz3
        dj += dz4 * X1
        dk += dz4 * Y1
        dl += dz4
        
        # 1 layer에 대한 역전파 계산
        dz1 = back_sigmoid(dz3, g, z1) + back_sigmoid(dz4, j, z1)
        dz2 = back_sigmoid(dz3, h, z2) + back_sigmoid(dz4, k, z2)
        da_ += dz1*x1
        db_ += dz1*x2
        dc_ += dz1*x3
        dd_ += dz1*x4
        de_ += dz1
        df_ += dz2*x1
        dg_ += dz2*x2
        dh_ += dz2*x3
        di_ += dz2*x4
        dj_ += dz2
    
    a_ -= eta*da_
    b_ -= eta*db_
    c_ -= eta*dc_
    d_ -= eta*dd_
    e_ -= eta*de_
    f_ -= eta*df_
    g_ -= eta*dg_
    h_ -= eta*dh_
    i_ -= eta*di_
    j_ -= eta*dj_
    
    g -= eta*dg
    h -= eta*dh
    i -= eta*di
    j -= eta*dj
    k -= eta*dk
    l -= eta*dl
    m -= eta*dm
    n -= eta*dn
    o -= eta*do
    p -= eta*dp
    q -= eta*dq
    r -= eta*dr
    s -= eta*ds
    t -= eta*dt
    u -= eta*du
    
print('\n',a_,b_,c_,d_,e_,f_,g_,h_,i_,j_,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u) 
```

```python
def print_z(x1, x2, x3, x4, y):
    z1 = node_output2(x1, x2, x3, x4, a_, b_, c_, d_, e_)
    z2 = node_output2(x1, x2, x3, x4, f_, g_, h_, i_, j_)
    X1 = sigmoid(z1)
    Y1 = sigmoid(z2)
    z3 = node_output(X1, Y1, g, h, i)
    z4 = node_output(X1, Y1, j, k, l)
    X2 = sigmoid(z3)
    Y2 = sigmoid(z4)
    z5 = node_output(X2, Y2, m, n, o)
    z6 = node_output(X2, Y2, p, q, r)
    X3 = sigmoid(z5)
    Y3 = sigmoid(z6)
    z = node_output(X3, Y3, s, t, u)
    
    return print('Pred:',round(z,0), ', Real:',y)

for i in range(len(test_data)):
    print_z(test_data[i][0], test_data[i][1], test_data[i][2], test_data[i][3], test_labels[i])
    
# 20개의 test data 중, 9개의 데이터를 바르게 분류한 것을 확인
# 직관적으로는 overrfiting이 발생했다고 판단하기 어려움
```
